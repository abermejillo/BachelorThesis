# Magnetic dipole measurements plotting

```python
import csv
from math import sqrt
import pandas as pd
import numpy as np

distance01 = []
B01 = []
with open('01mmb.csv',mode = 'r') as csv_file1:
    csv_reader1 = csv.reader(csv_file1,delimiter = ';')
    for row in csv_reader1:
        distance01.append(float(row[0].strip()))
        B01.append(float(row[1].strip()))

distance03 = []
B03 = []

with open('03mmb.csv',mode = 'r') as csv_file2:
    csv_reader2 = csv.reader(csv_file2,delimiter = ';')
    for row in csv_reader2:
        distance03.append(float(row[0].strip()))
        B03.append(float(row[1].strip()))
        
distance05 = []
B05 = []

with open('05mmb.csv',mode = 'r') as csv_file3:
    csv_reader3 = csv.reader(csv_file3,delimiter = ';')
    for row in csv_reader3:
        distance05.append(float(row[0].strip()))
        B05.append(float(row[1].strip()))

distance1 = []
B1 = []

with open('1mmb.csv',mode = 'r') as csv_file4:
    csv_reader4 = csv.reader(csv_file4,delimiter = ';')
    for row in csv_reader4:
        distance1.append(float(row[0].strip()))
        B1.append(float(row[1].strip()))
        
distance_sim = []
B_sim = []

with open('femmsimb.csv',mode = 'r') as csv_file5:
    csv_reader5 = csv.reader(csv_file5,delimiter = ';')
    for row in csv_reader5:
        distance_sim.append(float(row[0].strip()))
        B_sim.append(float(row[1].strip()))
        
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
import matplotlib.ticker as mticker
fig = plt.figure(figsize=(10,6))
ax = plt.axes(autoscale_on=True,xticks=(np.arange(-9, 9, step=2)),yticks=(np.arange(10, 40, step=5)))

x = np.linspace(-9,9,1000)
#ax.scatter(distance01,B01,s=100,marker="+",c="r")

ax.plot(distance_sim,B_sim,c='k',label="FEMM simulation")
ax.scatter(distance01,B01,s=70,marker=".",c="royalblue",label="0.1 mm")
ax.scatter(distance03,B03,s=70,marker=".",c="gray", label= "0.3 mm")
ax.scatter(distance05,B05,s=70,marker=".",c="darkorange",label= "0.5 mm")
ax.scatter(distance1,B1,s=80,marker=".",c="limegreen",label= "1 mm")


plt.legend(bbox_to_anchor=(1.05, 1),loc=0)


ax.yaxis.grid(True)
ax.grid(False)
#ax.legend(frameon=False)
#ax.axis('equal')
ax.set_xlim(-9, 9);
ax.set_ylim(15, 36);
ax.set_xlabel("$x$ (mm)")
ax.set_ylabel("$B$ (mT)")
#plt.xticks(range(0,75,5))
#for tick in ax.yaxis.get_major_ticks():
#    tick.tick1On = False
#    tick.tick2On = False
#ax.text(5, 74, ' $B_s=-0.0743+1.2452 B_g$ \n $R^2=0.99993$', fontsize=15,bbox={ 'alpha': 0.2, 'pad': 10})
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 16))
# Hide the top and right spines.
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.plot(1, 16, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_color('black')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
#plt.grid=
#ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
#ax.yaxis.set_major_locator(mticker.MaxNLocator(15))
#ax1.xaxis.set_major_locator(mticker.MaxNLocator(5))
plt.show()
fig.savefig('Femmsimprofile.svg')
```
