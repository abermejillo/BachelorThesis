# Plotting of magnetic profiles along the axis for simulations and experimental data

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

distance01 = []
B01 = []


with open('Sheet1.csv',mode = 'r') as csv_file1:
    csv_reader1 = csv.reader(csv_file1,delimiter = ',')
    for row in csv_reader1:
        distance01.append(float(row[0].strip()))
        B01.append(float(row[1].strip()))
distance11 = []
B11 = []
for i in range(len(distance01)):
    if (i%3)==1:
        distance11.append(distance01[i])
        B11.append(B01[i])
        
distance02 = []
B02 = []

with open('magnetic_axis_Ansys_1A.csv',mode = 'r') as csv_file2:
    csv_reader2 = csv.reader(csv_file2,delimiter = ',')
    for row in csv_reader2:
        distance02.append(float(row[0].strip()))
        B02.append(float(row[1].strip()))

distance03 = []
B03 = []

with open('Baxisfield_parametertweek8.csv',mode = 'r') as csv_file3:
    csv_reader3 = csv.reader(csv_file3,delimiter = ',')
    for row in csv_reader3:
        distance03.append(float(row[0].strip()))
        B03.append(float(row[1].strip()))

distance04 = []
B04 = []

with open('Baxis_field_nolateral.csv',mode = 'r') as csv_file4:
    csv_reader4 = csv.reader(csv_file4,delimiter = ',')
    for row in csv_reader4:
        distance04.append(float(row[0].strip()))
        B04.append(float(row[1].strip()))
errorbar = []        
for i in range(len(distance11)):
    errorbar.append(0.05)

fig1 = plt.figure(figsize=(10,6))
ax1 = plt.axes(autoscale_on=True)
ax1.scatter(distance11,B11,c='royalblue',label="B exp",s=30,marker='.')
ax1.errorbar(distance11,B11,yerr=errorbar,elinewidth=0.8)
ax1.plot(distance02,B02,c='r',label="B design")
ax1.plot(distance02,B03,c='g',label="B adapted")
ax1.yaxis.grid(True)
ax1.grid(False)
ax1.legend()
#ax.legend(frameon=False)
#ax.axis('equal')
ax1.set_xlim(-20,210);
ax1.set_ylim(0, 37);
ax1.set_xlabel("$y$ (mm)")
ax1.set_ylabel("$B$ (mT)")
plt.show()
fig1.savefig('comparison.svg')
```
